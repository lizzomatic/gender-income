## Introduction

As The Economist noted, there is some correlation between poverty and obesity in wealthy nations like the
US and European nations, such that poor people are more likely to be overweight than the wealthy. This
has often been explained by arguments that poorer people may struggle to afford healthy foods; be relegated
to processed or fast foods because they lack the time to prepare home meals or exercise because of long non
flexible work; or lack of education or knowledge about how to maintain a healthy weight. But the problem
with all of these explanations is that the correlation between income and weight in advanced countries is
driven almost entirely by women. As The Economist explains, “rich women are much thinner than poor
women but rich men are about as fat as poor men” (Economist, 2022).

## Data

Data is from the National Health Interview Survey (NHIS), a nationally-representative, cross-sectional
survey providing health and behavioral health information of the U.S. population beginning in 1963. Our
sample is constrained to the year 2018. It consists of men and women aged 18 and older who recorded working
at a job or business within the last week. BMI is measured as a person’s weight in kilograms divided by a
person’s height measured as meters squared. Just using an individual’s weight does not take into account
their height, and it can lead to an inaccurate depiction of body composition. Moreover, BMI is a greater
indicator of health status. Individuals with higher BMI are more subject to severe health limitations, and
therefore, they may face a negative impact on their earnings. Our EARNINGS variable is a catagorical variable, bracketed in \\$5,000 increments up to \\$25,000. Then is jumps to \\$10,000 increments up to \\$75,000. The final category is a catchall for any income above \$75,000.

**Hypothesis Test: Applied to the one specified predictor variable**

**Null hypothesis and its alternative for the hypothesis test of the slope coefficient.**

For Female:

**p-value with the relevant numbers for this specific analysis.**

**FUnder** has a statistically significant effect on earnings.
